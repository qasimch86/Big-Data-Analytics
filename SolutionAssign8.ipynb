```python
import pandas as pd
import numpy as np
df=pd.read_csv('Data.csv')
```

```python
df.head(2)
```

```python
target=pd.DataFrame(df.gnd)
df.drop(['Unnamed: 0','gnd'],axis =1,inplace=True)
```

```python
df.describe()
```

```python
# 2.	Apply any feature selection algorithm and choose the best features.
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
rfe=RFE(model,25)
fit = rfe.fit(df,target)
```

```python
fit.ranking_
```

```python
fit.support_
```

```python
df2=df[df.columns[fit.support_]]
```

```python
df2.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(df2,target['gnd'],train_size=0.7,random_state=7)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
lr=LogisticRegression()
lr.fit(X_train,Y_train)
pred=lr.predict(X_test)
```

```python
print('Train Accuracy :', lr.score(X_train,Y_train))
print('Test Accuracy :',lr.score(X_test,Y_test))
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=25)
pca.fit(df2)
```

```python
Comp_pca=pca.fit_transform(df2)
df_pca=pd.DataFrame(Comp_pca)
```
